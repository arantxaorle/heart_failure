```python
class ChinesePanda (object):

    def __init__(self, variables):
        self.config = variables
        self.weight = variables['PANDAS']['WEIGHT']
        self.height = variables['PANDAS']['HEIGHT']

    def BMI_CALCULATOR(self): 
        BMI = self.weight/self.height**2
        return BMI
        
```

```python
mypanda = ChinesePanda(config_variables)
```

```python
if config_variables['PANDAS']['CONTROL']:
    BMI=mypanda.BMI_CALCULATOR()
    print(BMI)
```
